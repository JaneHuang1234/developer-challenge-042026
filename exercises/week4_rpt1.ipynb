```python
from gen_ai_hub.proxy.native.sap.models import RPTRequest, PredictionConfig, TargetColumn
from gen_ai_hub.proxy.native.sap.client import RPTClient
```

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/noravth/developer-challenge-042026/refs/heads/main/sales.csv"
sales_data = pd.read_csv(url)
```

```python
rows = sales_data.to_dict(orient="records")
rows
```

```python
client = RPTClient()
prediction_config = PredictionConfig(target_columns=[TargetColumn(name = 'SALESGROUP', task_type='classification')])
body = RPTRequest(prediction_config=prediction_config, rows = rows)
response = client.predict(body = body, model_name = 'sap-rpt-1-small')
```
